### Introducing the data

While I could have used articles and books about how these topics correlate, I wanted to use this as an opportunity to look at only facts and data. Instead of relying solely on human opinions, I will be relying heavily on data and making graphs that use this data to conclude the impact of religion, conflict, and peace on the world economy. 

This data comes from the public data collection site *Data World* and was assembled by Kevin Nayer in 2016. It includes data from 156 countries and has their GDP per capita, Global Peace Index Score, Global Peace Ranking, and the percentage of the population that identifies as non-religious.

```python
import pandas as pd
import os

religionGPI = os.path.join("InputData", "religionGPI.csv") 
religion_gpi_data = pd.read_csv(religionGPI)
print(religion_gpi_data)
```

### Graphing the Data 

Using the data above I can graph many different charts to depict the correlation between religious population, GDP, and GPI. 

```python
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure
import numpy as np

x = religion_gpi_data["gpi_rank"]
y = religion_gpi_data["percentage_non_religious"]

m, b = np.polyfit(x, y, 1)
figure(figsize=(10, 6), dpi=80)

plt.figure
plt.scatter(x, y)
plt.plot(x, m*x + b, color="red", label="Line of Best Fit")
plt.title("Global Peace Ranking vs. Percentage of Non-Religious Pop")
plt.xlabel("GPI Ranking")
plt.ylabel("% Non-Religious Pop")
plt.legend(loc="upper right")
plt.savefig("GPIvsReliPop.png")
```

### Interpreting Graph #1

We can see here is that there is a correlation between GPI Ranking and the non-religious population of a country. Each dot represents one country from the dataset. On the x-axis, we have a GPI ranking from 1 to 160, with one being the most peaceful and 160 being the least peaceful, and on the y-axis, we have the % population that identifies as non-religious. We can see that as the non-religious population decreases, the level of peace also decreases. There are a few exceptions to this rule. On the righthand side of the graph, there is one country with a GPI ranking of 153 and a non-religious population of 32%. This country is Russia; I predict that this has more to do with their style of government and less so with the levels of religion among the population.  

```python
x = religion_gpi_data["gpi_score"]
y = religion_gpi_data["gdp_pc"]

m, b = np.polyfit(x, y, 1)
figure(figsize=(10, 6), dpi=80)

plt.figure
plt.scatter(x, y)
plt.plot(x, m*x + b, color="red", label="Line of Best Fit")
plt.title("Global Peace Index vs. GDP Per Capita")
plt.xlabel("GPI")
plt.ylabel("GDP (per capita)")
plt.legend(loc="upper right")
plt.savefig("GPIvsGDP.png")
```

### Interpreting Graph #2 

In the graph above, I am comparing each country's global peace index (1.0 being most peaceful, 3.5 being least peaceful) and their GDP per capita. As I expected, as a country becomes less peaceful, its GDP per capita also decreases. Given the graphs above, I conclude that GDP, GPI, and the religious population of a country are all correlated with one another.

## Looking at Syria

Thus far, we have looked at economics on a global level. By taking a closer look at just one country over time we can see how various levels of religious freedom and civil liberties can change a countries economy. 

I chose to look at Syria because it has a history of intertwined politics and religious affairs. In the 2000's Syria also saw times of economic success compared to most of its neighbors, despite having relatively little resources. According to William R. Polk from The Atlantic, Syrians had a GPD per capita of $5,000, 90% of school children attended primary or secondary school, and roughly 9/10 Syrians were literate. However, this all came to a halt when the drought and civil war ravaged the country. 

The civil war in Syria started as civilians protesting their government's lack of help during the drought and subsequent famine but quickly turned into a political and religious fight.

The cost of this war on the lives of Syrians has been detrimental. In 2013, only a few years after the start of the war, it was estimated to have cost Syris upwards of $150 billion, with more than 2 million people fleeing the state. (Polk, 2013) From 2007 to 2019, Syria saw a 222.3% increase in the cost of violence in the region. (EVP, 2021) Furthermore, the effects have spilled over to the neighboring countries of Lebanon, Jordan, Iraq, and Isreal. These countries are mainly affected by the increase in refugees. 

This conflict, caused by a history of turmoil and fueled by religious motives, has resulted in an economic impact of which we do not know how long it will take to recover. The good news is that the recovery has already started. The Syrian conflict is no longer at the height it once was and the economic impact of the violence has been on the decline since 2015.
